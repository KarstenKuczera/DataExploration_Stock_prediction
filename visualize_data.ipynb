```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.25,rc={'figure.figsize':(10,8)})
```

```python
def plot_next_day(ticker, label, data):
    columns_epochs = ["50", "500", "1000"]

    fig, ax = plt.subplots()
    plt.title(f"{ticker} predict next day")
    bars = ax.bar(columns_epochs, data[0:3])
    ax.bar_label(bars)
    ax.set_xlabel("Epoch")
    ax.set_ylabel(f"{label}")
    plt.show()
```

```python
def plot_15_days(ticker, label, data):
    columns_epochs = ["50", "500", "1000"]

    fig, ax = plt.subplots()
    plt.title(f"{ticker} predict 15 Days ahead")
    bars = ax.bar(columns_epochs, data[3:7])
    ax.bar_label(bars)
    ax.set_xlabel("Epoch")
    ax.set_ylabel(f"{label}")
    plt.show()
```
