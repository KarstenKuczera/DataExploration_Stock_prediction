#### Vorteile
* Gute Dokumentation
* Ausgabe direkt in einem DataFrame
* Es können mehrere Aktien gleichzeitig ausgegeben werden
* Kurse in Minutenabständen möglich

#### Nachteile 
* Minutenabstände nur in einem Zeitraum von maximal 7 Tagen auslesbar
* Minutenabstände nur von Zeiträumen in den letzten 30 Tagen auslesbar
* Keinen Generellen Informationen zu den Unternehmen

```python
import yfinance as yf  
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import seaborn as sns
sns.set(font_scale=1.25)
```

```python
data = yf.download('BTC-USD','2021-01-01','2021-09-30')
data.head()
```

```python
data = yf.download(['BAS.DE','AMD'],'2021-01-01','2021-09-30')
data["Close"].plot(figsize=(16, 14))
plt.show()
```

```python
x = datetime.today() - timedelta(days=6)

data = yf.download(tickers='BAS.DE', start= x.strftime('%Y-%m-%d'), interval='1m')
data
```
